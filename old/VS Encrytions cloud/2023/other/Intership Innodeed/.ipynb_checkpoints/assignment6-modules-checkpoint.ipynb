```python
"""Write a Python program that takes a user-defined class and a value as input. The program should determine 
if the given value is a method of the class using the types.MethodType() function from the types module. 
The program should return a boolean value indicating whether the value is a method of the class or not."""

import types

def is_method_of_class(user_class, value):
    return isinstance(value, types.MethodType) and value.__self__ == user_class

# Define a user-defined class
class MyClass:
    def my_method(self):
        pass

# Test the program
my_class = MyClass()
value = my_class.my_method
is_method = is_method_of_class(MyClass, value)
print("Is value a method of the class?", is_method)


```

```python
""""Develop a Python program that analyzes a directory containing multiple Python files and generates a report 
summarizing the imported modules in each file. The program should recursively traverse the directory, identifying 
all Python files (files with the .py extension) and examining their contents to extract the imported modules. 
The report should list each Python file along with the imported modules in that file. Additionally, the program 
should include statistics such as the total number of files processed, the total number of imported modules
across all files, and the frequency of each imported module. The report should be saved to a text file for 
future reference."""

import os
import re
from collections import defaultdict

def extract_imported_modules(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
        import_regex = r'^\s*import\s+(\S+).*$'
        from_import_regex = r'^\s*from\s+(\S+)\s+import.*$'
        imports = re.findall(import_regex, content, re.MULTILINE)
        from_imports = re.findall(from_import_regex, content, re.MULTILINE)
        return imports + from_imports

def analyze_directory(directory_path):
    file_count = 0
    imported_modules = defaultdict(int)

    report = ""
    for root, dirs, files in os.walk(directory_path):
        for file_name in files:
            if file_name.endswith('.py'):
                file_count += 1
                file_path = os.path.join(root, file_name)
                imported_modules_list = extract_imported_modules(file_path)

                report += f"File: {file_path}\n"
                report += "Imported Modules:\n"
                for module in imported_modules_list:
                    imported_modules[module] += 1
                    report += f"- {module}\n"
                report += "\n"

    # Write summary statistics
    report += "Summary Statistics:\n"
    report += f"Total Files Processed: {file_count}\n"
    report += f"Total Imported Modules: {sum(imported_modules.values())}\n"
    report += "Imported Module Frequencies:\n"
    for module, frequency in imported_modules.items():
        report += f"- {module}: {frequency}\n"

    # Save report to a text file
    report_file_path = os.path.join(directory_path, 'import_report.txt')
    with open(report_file_path, 'w') as report_file:
        report_file.write(report)

    print(f"Report generated and saved to {report_file_path}")
    with open(report_file_path, 'r') as report_file:
        content = report_file.read()
        print(content)

# Test the program
directory_path = input("Enter the directory path: ")
analyze_directory(directory_path)

```

```python
"""Write a Python program that dynamically imports a module specified by user input. The program should prompt
the user to enter the name of the module they want to import, then load and execute that module. Afterward, 
the program should display a list of all the functions defined in the imported module."""


def import_and_execute_module():
    module_name = input("Enter the name of the module to import: ")

    try:
        imported_module = __import__(module_name)
        module_functions = [func for func in dir(imported_module) if callable(getattr(imported_module, func))]

        print(f"Functions defined in the module '{module_name}':")
        for func in module_functions:
            print("- " + func)

    except ImportError:
        print(f"Failed to import module '{module_name}'")
import_and_execute_module()

```

```python
"""Create a Python program that dynamically imports all modules from a given directory and identifies any circular
dependencies between the modules. The program should detect if any modules have circular imports and display a 
list of the affected modules."""

import os
import importlib
from collections import defaultdict

def get_imported_modules(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    imported_modules = set()
    for line in lines:
        if line.startswith("import") or line.startswith("from"):
            words = line.split()
            if len(words) > 1:
                imported_module = words[1].split('.')[0]
                imported_modules.add(imported_module)

    return imported_modules

def find_circular_dependencies(directory_path):
    modules = set()
    circular_dependencies = defaultdict(set)

    for root, dirs, files in os.walk(directory_path):
        for file_name in files:
            if file_name.endswith('.py'):
                file_path = os.path.join(root, file_name)
                module_name = os.path.splitext(file_name)[0]

                try:
                    module = importlib.import_module(module_name)
                    modules.add(module)
                except Exception:
                    circular_dependencies[module_name].add(module_name)
                    continue

                imported_modules = get_imported_modules(file_path)

                for imported_module in imported_modules:
                    try:
                        imported_module = importlib.import_module(imported_module)
                        modules.add(imported_module)
                    except Exception:
                        circular_dependencies[module_name].add(imported_module)

    return circular_dependencies

# Prompt the user for the directory path
directory_path = input("Enter the directory path: ")

# Find circular dependencies
circular_dependencies = find_circular_dependencies(directory_path)

# Display the affected modules
if circular_dependencies:
    print("Circular dependencies found:")
    for module, dependencies in circular_dependencies.items():
        print(f"- Module: {module}")
        print(f"  Dependencies: {', '.join(dependencies)}")
else:
    print("No circular dependencies found.")


```

```python
"""Implement a Python program that measures the execution time of a specific function from a module. 
The program should prompt the user to enter the name of the module and the function they want to measure. 
After execution, the program should display the elapsed time taken by the function."""


import time
import importlib

def measure_execution_time(module_name, function_name, *args, **kwargs):
    try:
        module = importlib.import_module(module_name)
        function = getattr(module, function_name)
    except (ImportError, AttributeError):
        print("Error: Failed to import module or find function.")
        return

    # Convert input arguments to appropriate types
    converted_args = []
    for arg in args:
        try:
            converted_arg = eval(arg)  # Evaluate string representation to actual value
        except (NameError, SyntaxError):
            converted_arg = arg  # Use the argument as is if evaluation fails
        converted_args.append(converted_arg)

    start_time = time.time()
    function(*converted_args, **kwargs)
    end_time = time.time()

    elapsed_time = end_time - start_time
    print(f"Elapsed time: {elapsed_time} seconds")

# Prompt user for module and function names
module_name = input("Enter the name of the module: ")
function_name = input("Enter the name of the function: ")

# Prompt user for function arguments (optional)
args = input("Enter function arguments (comma-separated): ").split(",")
kwargs = {}

# Measure execution time
measure_execution_time(module_name, function_name, *args, **kwargs)
```

```python
"""Create a Python program that detects and reports any unused imports within a module. The program should analyze
the module's source code, identify imported modules that are not used, and display a list of the unused imports."""

import ast

def find_unused_imports(module_path):
    with open(module_path, 'r') as file:
        source_code = file.read()

    tree = ast.parse(source_code)
    import_names = set()
    used_imports = set()

    # Find all import names
    for node in ast.walk(tree):
        if isinstance(node, ast.Import):
            for alias in node.names:
                import_names.add(alias.name)
        elif isinstance(node, ast.ImportFrom):
            for alias in node.names:
                import_names.add(alias.name)

    # Find used import names
    for node in ast.walk(tree):
        if isinstance(node, ast.Name) and isinstance(node.ctx, ast.Load):
            used_imports.add(node.id)

    # Find unused import names
    unused_imports = import_names - used_imports

    return unused_imports

# Test the program
module_path = "C:\\Users\\harsh\\OneDrive\\Desktop\\Python\\armstrong.py"
unused_imports = find_unused_imports(module_path)

print("Unused Imports:")
for unused_import in unused_imports:
    print(unused_import)
```

```python
"""Develop a Python program that finds the longest chain of dependencies between modules in a given Python project.
The program should analyze the import statements within each module and identify the longest sequence of module 
dependencies."""

import os
import re

def extract_imported_modules(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
        import_regex = r'^\s*import\s+(\S+).*$'
        from_import_regex = r'^\s*from\s+(\S+)\s+import.*$'
        imports = re.findall(import_regex, content, re.MULTILINE)
        from_imports = re.findall(from_import_regex, content, re.MULTILINE)
        return imports + from_imports

def build_dependency_graph(directory_path):
    dependency_graph = {}

    for root, dirs, files in os.walk(directory_path):
        for file_name in files:
            if file_name.endswith('.py'):
                file_path = os.path.join(root, file_name)
                imported_modules = extract_imported_modules(file_path)
                module_name = os.path.splitext(file_name)[0]
                dependency_graph[module_name] = imported_modules

    return dependency_graph

def dfs(node, graph, visited, current_path, longest_path):
    visited.add(node)
    current_path.append(node)

    for neighbor in graph.get(node, []):
        if neighbor not in visited:
            dfs(neighbor, graph, visited, current_path, longest_path)

    if len(current_path) > len(longest_path):
        longest_path.clear()
        longest_path.extend(current_path)

    current_path.pop()
    visited.remove(node)

def find_longest_dependency_chain(directory_path):
    dependency_graph = build_dependency_graph(directory_path)
    visited = set()
    longest_path = []

    for module in dependency_graph:
        if module not in visited:
            dfs(module, dependency_graph, visited, [], longest_path)

    return longest_path

# Test the program
directory_path ="C:\\Users\\harsh\\OneDrive\\Desktop\\Python\\"
longest_chain = find_longest_dependency_chain(directory_path)

print("Longest Dependency Chain:")
print(" -> ".join(longest_chain))
```

```python
"""Write a Python program that implements a custom logging module. The module should provide functionality similar to the built-in logging module, but with additional features.
Your custom logging module should have the following requirements:
It should support different log levels such as DEBUG, INFO, WARNING, ERROR, and CRITICAL.
It should allow logging messages to be written to a file, with the ability to specify the file name and log format.
"""

import logging
class LoggerDemoCon:
   def logging_functionality(self):
       logger = logging.getLogger('demologger')
       logger.setLevel(logging.INFO)
       consoleHandler = logging.StreamHandler()
       consoleHandler.setLevel(logging.INFO)
       formatter = logging.Formatter('%(asctime)s - %(name)s%(levelname)s: %(message)s', datefmt='%m/%d/%Y %I:%M:%S%p')
       consoleHandler.setFormatter(formatter)
       logger.addHandler(consoleHandler)
       logger.debug('debug message')
       logger.info('info message')
       logger.warn('warn message')
       logger.error('error message')
       logger.critical('critical message')
demo = LoggerDemoCon()
demo.logging_functionality()
```

```python
"""Write a Python program that makes an HTTP GET request to a given URL and retrieves the response data. 
However, the program should use the requests module to handle the network communication."""

import requests

def make_get_request(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.text
    except requests.exceptions.RequestException as e:
        print("Error occurred during the request:", e)
        return None
url = "https://www.google.com"
response_data = make_get_request(url)

if response_data is not None:
    print("Response data:")
    print(response_data)
```

```python
"""Write a Python program to round a decimal value to the nearest multiple of 0.004, unless already an exact 
multiple of 0.0005. Use decimal.Decimal"""

from decimal import Decimal

def round_to_10_cents(x):
    remainder = x.remainder_near(Decimal('0.10'))
    if abs(remainder) == Decimal('0.0005'):
        return x
    else:
        return x - remainder

for x in range(80, 120):
    y = Decimal(x) / Decimal('1E2')
    print("{0} rounds to {1}".format(y, round_to_10_cents(y)))
```
