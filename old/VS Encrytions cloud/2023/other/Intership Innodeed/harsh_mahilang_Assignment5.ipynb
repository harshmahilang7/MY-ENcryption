```python
"""Write a python code for explaining the inheritance in python.
Write code for all type of python inheritance with proper code example."""

print()
print("simple inheritance")
class single_base:
    def display_base(self):
        print("i am base class")
    def display_child(self):
        print("i am simple inheritance form base class")
class child(single_base):
    def display_im(self):
        print("i am child i can call base method and attributes")
    
d=child()
d.display_im()
d.display_child()

print()
print("multilevel inheritance")

class garand_father:
    def show_garand_father(self):
        print("i am garand_father")
class father(garand_father):
    def show_father(self):
        print("i am father")
        
class son(father):
    def show_son(self):
        print("i am son")
ml=son()
ml.show_garand_father()
ml.show_father()
ml.show_son()

print()
print("multiple inheritance")
class base1:
    def show_base1(self):
        print("i am base1")
class base2:
        def show_base2(self):
            print("i am base2")
class base3:
        def show_base3(self):
            print("i am base3")
class derived(base1,base2,base3):
    pass
mul_in=derived()
mul_in.show_base1()
mul_in.show_base2()
mul_in.show_base3()

print()
print("hierarchical inheritance")
class parent:
    def show_parent(self):
        print("i am parent call form child 1")
    def show_parent1(self):
        print("i am parent call form child 2")
class child1(parent):
    pass
class child2(parent):
    pass
obj1=child1()
obj1.show_parent()
obj2=child2()
obj2.show_parent1()


print()
print("hybrid inheritance")
class A:
    def show_A(self):
        print("i am class A")

class B(A):
        def show_B(self):
            print("i am class B")

class C(A):
        def show_C(self):
            print("i am class C")

class D(B,C):
        def show_D(self):
            print("i am class D")
        
obj=D()
obj.show_D()
obj.show_C()
obj.show_B()
obj.show_A()
```

```python
"""Give an example of encapsulation in python that refers to Loan Class"""

class Loan:
    def __init__(self, amount, interest_rate):
        self.__amount = amount
        self.__interest_rate = interest_rate

    def get_amount(self):
        return self.__amount

    def set_amount(self, amount):
        self.__amount = amount

    def get_interest_rate(self):
        return self.__interest_rate

    def set_interest_rate(self, interest_rate):
        self.__interest_rate = interest_rate

my_loan = Loan(27000, 2)

print("Initial Loan Amount:", my_loan.get_amount())
print("Initial Interest Rate:", my_loan.get_interest_rate())

my_loan.set_amount(30000)
my_loan.set_interest_rate(3)

print("Modified Loan Amount:", my_loan.get_amount())
print("Modified Interest Rate:", my_loan.get_interest_rate())

```

```python
"""Develop a Rectangle class that accepts the length and width as parameters during initialization. Include 
methods within the class to compute and return the area and perimeter of the rectangle. Validate the class functionality 
by creating an instance of the Rectangle with specific dimensions and displaying the calculated area and perimeter."""

class Rectangle:
    def __init__(self,l,w):
        self.Length=l
        self.Width=w
    def perimeter(self):
        per=2*(self.Length+self.Width)
        return("the perimeter of the rectangle=",per)
    def area(self):
        area=self.Length*self.Width
        return("the area of the rectangle=",area)

obj1=Rectangle(80,40)
print(obj1.area())
print(obj1.perimeter())

    
```
